```python
#Create DataFrame
import pandas as pd
import numpy as np
df = pd.DataFrame(
    {'integer':[1,2,3,6,7,23,8,3],
     'float':[2,3.4,5,6,2,4.7,4,8],
     'string':['saya','aku', np.NaN ,'cinta','kamu','a','b','indonesia']}
)
```

```python
#show data in DataFrame
df
```

```python
#Show data based on index
df.loc[1]
```

```python
#show data based on columns selected
df[['string','float']]
```

```python
#show data with condition
df[df['float']>4]
```

```python
#rename column in DataFrame
df.rename(columns={'string':'char'})
```

```python
#Show DataFrame after rename column
df
```

```python
#Show Na value in DataFrame
df.isnull()
```

```python
df.dropna() 
```

```python
#Show mean, median, and maximum in Data Frame
mean = df['float'].mean()
print("mean =", mean)  
median = df['float'].median()
print("median = ", median) 
max = df['float'].max()
print("max =", max)
```
